###**1.8 Legend :-**

A legend is an area describing the elements of the graph. In the matplotlib library, theres a function called legend() which is used to Place a legend on the axes.

The attribute Loc in legend() is used to specify the location of the legend.Default value of loc is loc=best (upper left). The strings upper left, upper right, lower left, lower right place the legend at the corresponding corner of the axes/figure.

The attribute bbox_to_anchor=(x, y) of legend() function is used to specify the coordinates of the legend, and the attribute ncol represents the number of columns that the legend has.Its default value is 1.

####**1.8.1 Method-1**

````python
x = np.arange(7)
y1 = np.random.randint(1,100,7)
y2 = np.random.randint(1,100,7)

d1 = {"family":"serif",
      "color" : "red",
      "size" : 20}

d2 = {"family":"sans-serif",
      "color" : "green",
      "size" : 15}

plt.figure(figsize=(8,5))

plt.plot(x,y1,color="purple",ls="dashdot",marker="o",ms=6,mec="red",mfc="yellow",label="1st value")
plt.plot(x,y2,color="blue",ls="dotted",marker=".",ms=6,mec="black",mfc="white",label="2nd value")
plt.title("Simple Line Plot", fontdict = d1)
plt.xlabel("Random X Values",fontdict = d2)
plt.ylabel("Random Y Value",fontdict = d2)
plt.grid()
plt.legend()
plt.show()
````

##**2 Bar Chart:-**

A bar chart is a way of summarizing a set of categorical data (continuous data can be made categorical by auto-binning). The bar chart displays data using a number of bars, each representing a particular category. The height of each bar is proportional to a specific aggregation (for example the sum of the values in the category it represents). The categories could be something like an age group or a geographical location. It is also possible to color or split each bar into another categorical column in the data, which enables you to see the contribution from different categories to each bar or group of bars in the bar chart.

###**2.1 Simple Barchart :-**

````python
x = np.array(["Python","C","C++","Java"])
y = np.array([85,70,75,80])

d1 = {"family": "Serif",
      "color" : "red",
      "size" : 20}

d2 = {"family": "Monospace",
      "color" : "green",
      "size" : 15}

c = np.array(["yellow","blue","green","purple"])
plt.bar(x,y,color=c,width=0.4,align="center",alpha=0.6)
plt.xlabel("Languages",fontdict=d2)
plt.ylabel("Popularity",fontdict=d2)
plt.title("Bar Plot",fontdict=d1)
plt.show()
````

````python
name = np.array(["Mihir","Vaibhav","Dev","Raj","Rahul"])

marks = np.array([98,96,92,86,92])

plt.figure(figsize=(6,4.5))

plt.bar(name,marks,color="green",width=.5)
plt.xlabel("Names")
plt.ylabel("CN Marks")
plt.title("CE Department")
plt.yticks(np.arange(0,101,10))
plt.show()
````

###**2.2 Stack Barplot :-**

In a stacked barplot, subgroups are displayed on top of each other. The code is very similar with the previous post #11-grouped barplot.

Instead of passing different x axis positions to the function, you will pass the same positions for each variable. Additionally, in order to draw bars on top of each other, you should use bottom parameter of the bar() function.

This parameter will set the bottom value (bottom line) of the bar.

````python
depart = np.array(["CE","EE","IC","IT"])
year = np.array(["2020","2021","2022"])
ce = np.array([74,86,98])
ee = np.array([47,49,53])
ic = np.array([37,42,38])
it = np.array([61,67,62])

d1 = {"family": "Serif",
      "color" : "red",
      "size" : 20}

d2 = {"family": "Monospace",
      "color" : "green",
      "size" : 15}

start_ce = np.zeros(len(year),dtype=int)
start_ee = ce + start_ce
start_ic = ee + start_ee
start_it = ic + start_ic

plt.figure(figsize=(6,5.5))

plt.bar(year,ce,bottom=start_ce,label=depart[0],width=0.4,color="#87CEEB")
plt.bar(year,ee,bottom=start_ee,label=depart[1],width=0.4,color="blue")
plt.bar(year,ic,bottom=start_ic,label=depart[2],width=0.4,color="#FFC0CB")
plt.bar(year,it,bottom=start_it,label=depart[3],width=0.4,color="purple")
plt.legend(title="Depart.",fancybox=True)
plt.ylim(0,400)
plt.xlabel("Year",fontdict=d2)
plt.ylabel("Number of Student Placed",fontdict=d2)
plt.title("VGEC Placements",fontdict=d1)
plt.yticks(np.arange(0,350,25))
plt.show()
````

````python
year = np.array(["2018","2019","2020","2021","2022"])
cse_plac = np.array([75,82,73,79,80])
it_plac = np.array([69,63,51,68,74])
ec_plac = np.array([47,38,29,41,57])
ee_plac = np.array([53,47,51,63,70])

start_ec = cse_plac + it_plac
start_ee = start_ec + ec_plac

d1 = {"family" : "serif",
     "color" : "red",
     "size" : 25}

d2 = {"family" : "cursive",
     "color" : "green",
     "size" : 15}
plt.figure(figsize=(6,4.5))
plt.bar(year,cse_plac,width=.4,color="#00BFFF",label="CE")
plt.bar(year,it_plac,bottom=cse_plac,width=.4,color="#D3D3D3",label="IT")
plt.bar(year,ec_plac,bottom=start_ec,width=.4,color="orange",label="EC")
plt.bar(year,ee_plac,bottom=start_ee,width=.4,color="purple",label="EE")
plt.xlabel("Year",fontdict=d2)
plt.ylabel("Number of Students Placed",fontdict=d2)
plt.title("VGEC Placement 2021",fontdict=d1)
plt.yticks(np.arange(0,301,25))
plt.legend()
plt.show()
````

###**2.3 Grouped Barchart :-**

To plot a Grouped Bar Chart using Matplotlib, create a subplot using subplots() function, and in this subplot call bar() function with different X-axis position to draw each of the bar graph from different individual bar graphs, so that they form groups

````python
male  = np.array([74,80,89])
female = np.array([63,86,74])
year = ["2020","2021","2022"]
d1 = {"family": "Serif",
      "color" : "red",
      "size" : 20}

d2 = {"family": "Monospace",
      "color" : "green",
      "size" : 15}

w = 0.2

start_1 = np.arange(len(year),dtype="i")
start_2 = start_1 + w

plt.figure(figsize=(6,5.5))

plt.bar(start_1,male,width=w,label="Male",color="#87CEEB")
plt.bar(start_2,female,width=w,label="Female",color="blue")
plt.legend()
plt.ylim(0,120)
plt.yticks(np.arange(0,101,10))
plt.xticks(start_1+0.1,year)
plt.xlabel("Year",fontdict=d2)
plt.ylabel("Number of Employees Joined",fontdict=d2)
plt.title("TechSolution PVT.LTD.",fontdict=d1)
plt.show()
````

##**3 Pie Chart :-**

A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data. The slices of pie are called wedges. The area of the wedge is determined by the length of the arc of the wedge.

The area of a wedge represents the relative percentage of that part with respect to whole data. Pie charts are commonly used in business presentations like sales, operations, survey results, resources, etc as they provide a quick summary.

Matplotlib API has pie() function in its pyplot module which create a pie chart representing the data in an array.

###**3.1 Parameters to the pie() :-**

labels = that is used for giving the labels to the pie chart elements

startangle = default t is 0 , but we can set it according to our choice and it will draw graph anticlock-wise

explode = used for extract some portion of pie chart out of it , default it is 0

shadow = it is used for shadow the portion of the chart , default it is false

colors = It is used for providing different color for the elements of pie chart

autopct = It is used for give the percentage to the elements (e.g. autopct ="%2.2f%%" - It will add the two digit before decimal point and one digit after decimal point and shows for example 21.7% on the element)

###**3.2 Simple Pie Chart :-**

````python
student_cat = np.array(["Excellent","Good","Average","Poor"])
students = np.array([13,19,32,7])

plt.figure(figsize=(10,7.5))

plt.pie(students,labels=student_cat,autopct="%2.2f%%")
plt.legend(title="Performance")
plt.show()
````

````python
x = np.array([10,20,30,40])
label = np.array(["c","c++","java","python"])
ex = np.array([0.2,0.0,0.0,0.0])
plt.figure(figsize=(10,7.5))
plt.pie(x,labels=label,rotatelabels=True,autopct="%0.2f%%",explode=ex)
plt.show()
````

##**4 Histrogram :-**

A histogram is basically used to represent data provided in a form of some groups.It is accurate method for the graphical representation of numerical data distribution.It is a type of bar plot where X-axis represents the bin ranges while Y-axis gives information about frequency.

To create a histogram the first step is to create bin of the ranges, then distribute the whole range of the values into a series of intervals, and count the values which fall into each of the intervals.Bins are clearly identified as consecutive, non-overlapping intervals of variables.The matplotlib.pyplot.hist() function is used to compute and create histogram of x.

The following table shows the parameters accepted by matplotlib.pyplot.hist() function

###**4.1 Simple Histogram :-**

````python
student_marks = np.random.randint(0,101,50)
print(student_marks)

grade = np.array([0,33,65,75])
plt.figure(figsize=(8,5.5))

plt.hist(student_marks,color="#AE32BC")
plt.xticks(grade)
plt.show()
````

##**5 Scatter Plot :-**

With Pyplot, you can use the **scatter()** function to draw a scatter plot.

The scatter() function plots **one dot** for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis

###**5.1 Simple Scatter Plot :-**

````python
x = np.array([12,4,5,6,7])
y = np.array([11,4,5,6,9])

plt.figure(figsize=(6,4.5))

plt.scatter(x,y,marker="o",color=["green","black","blue","orange","brown"],s=[100,60,40,90,70],alpha=0.3)
plt.xticks(np.arange(13))
plt.yticks(np.arange(14))
plt.grid()
plt.show()
````

````python
x1 = np.array([7,5,2,9,4])
y1 = np.array([3,4,8,7,1])

x2 = np.array([8,2,3,1,7])
y2 = np.array([4,5,9,7,3])

plt.figure(figsize=(10,6))

plt.scatter(x1,y1,s=15,alpha=0.7,color="orange",label="smoker")
plt.scatter(x2,y2,s=20,alpha=0.9,color="#87CEEB",label="non-smoker")
plt.legend()
plt.xticks(np.arange(11))
plt.yticks(np.arange(11))
plt.ylim(0,13)
plt.xlabel("Year 2022")
plt.ylabel("Smokers")
plt.title("Smoking Detail")
plt.show()
````

###**5.2 Cmap with Scatter Plot :-**

````python
x = np.array([10,40,70,30,65])
y = np.array([400,750,630,350,70])

colors = np.array([70,90,40,75,20])
sizes = np.array([100,250,50,300,450])

plt.figure(figsize=(10,8))

plt.scatter(x,y,c=colors,cmap="rainbow",s=sizes)
plt.colorbar()
plt.ylim(0,800)
plt.show()
````
